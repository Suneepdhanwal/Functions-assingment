Ques.1 What is the difference between a function and a method in Python?

Ans.1

Function

A function is a self-contained unit of code designed to perform a specific task. It is defined independently and can be called from anywhere in the program, without being associated with a particular object. Functions promote code reusability and modularity.

Method

A method, on the other hand, is a function that is associated with an object. It operates on the data of that object and can only be called through that object. Methods define the behavior of objects and are fundamental to object-oriented programming.

```python
#example # Function
def add(x, y):
  return x + y

result = add(5, 3) # Calling the function
print(result)

#method
class Dog:
    def my_method(self):
        print("I am a Dog")

dog = Dog()
dog.my_method()  # Prints "I am a Dog"
```

Ques.5 What are iterators in Python and how do they differ from iterables?

Ans.5
Iterables

An iterable is any Python object capable of returning its members one at a time. It's like a collection of items that you can loop through. Examples of iterables include lists, tuples, strings, dictionaries, and sets.

Iterators

An iterator is an object that represents a stream of data; it returns the elements of the iterable one at a time. All iterators are iterables, but not all iterables are iterators. An iterator is created by calling the __iter__() method on an iterable, and then the __next__() method to get each element.

```python
#Iterables Example
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

#Iterators Example
fruits = ["apple", "banana", "cherry"]
it = iter(fruits)  # Creating an iterator

print(next(it))  # Output: apple
print(next(it))  # Output: banana
print(next(it))  # Output: cherry
# print(next(it))  # Raises StopIteration exception
```

Ques.6  Explain the concept of generators in Python and how they are defined.

Ans.6 Generators in Python are a special type of iterable that allow you to iterate over a sequence of values lazily, meaning they generate values on the fly and only when needed. This can be very memory efficient, especially when dealing with large datasets.

Generators are defined using either generator functions or generator expressions.
1. Generator Functions:-
A generator function is defined like a normal function but uses the yield statement to return values one at a time. Each time yield is called, the function's state is saved, and it can be resumed later from where it left off.

2. Generator Expressions
Generator expressions are similar to list comprehensions but use parentheses instead of square brackets. They provide a concise way to create generators.

```python
#Example
#Generator Functions
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

# Using the generator function
counter = count_up_to(5)
for number in counter:
    print(number)

#Generator Expression
squares = (x * x for x in range(10))
for square in squares:
    print(square)
```

Ques.7 What are the advantages of using generators over regular functions?

Ans.7

Memory Efficiency:-  Generators use less memory by generating values on the fly.

Lazy Evaluation:-  Values are produced only when needed, saving computation time.

Improved Performance:-  Iteration can start immediately without pre-computation.

State Retention:-  Generators keep track of their position automatically.

Simpler Code:-  More readable and maintainable code with fewer auxiliary structures.

Infinite Sequences:-  Generators can handle infinite sequences effortlessly.

```python
#Example
#Memory efficiency
def generate_numbers():
    for i in range(1000000):
        yield i

gen = generate_numbers()
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
# and so on, without holding all values in memory

#Lazy Evaluation
def generate_large_numbers():
    for i in range(1, 1000000):
        yield i * i

gen = generate_large_numbers()
# Computation of squares happens as needed

#Simpler Code with State Retention
def countdown(n):
    while n > 0:
        yield n
        n -= 1

cd = countdown(5)
for num in cd:
    print(num)
# Output:
# 5
# 4
# 3
# 2
# 1
```

Ques.9 Explain the purpose and usage of the `map()` function in Python.

Ans.9 **Purpose**: The map() function applies a given function to all items in an iterable and returns a map object.

**Syntax**: map(function, iterable, ...)

**Usage**: Transform data by applying a function to each element of an iterable. It can work with multiple iterables and support lambda functions for concise definitions.

```python
#Example
def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]
```

Ques.10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans.10 Key Differences

map(): Transforms each element in the iterable using a given function and returns an iterable of the transformed elements.

reduce(): Reduces the iterable to a single cumulative value by applying a function that takes two arguments.

filter(): Filters elements based on a condition defined by a function that returns true or false.

```python
#Example
#Map()
numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x * x, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16]

#Reduce()
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24

#Filter()
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4, 6]
```

```python
#Ques.1  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

#Ans.1
def sum_of_even_numbers(numbers):
    even_sum = 0
    for number in numbers:
        if number % 2 == 0:
            even_sum += number
    return even_sum

# Example usage:
numbers = [47, 11, 42, 13, 24, 36]
result = sum_of_even_numbers(numbers)
print(result)  # Output: 102
```

```python
#Ques.5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
#Ans.5 class FibonacciIterator:
class Squares:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        self.count += 1
        return self.count ** 2

# Example usage:
num_terms = 10
squares_seq = Squares(num_terms)
for num in squares_seq:
    print(num)
```

```python
#Ques.6 Write a generator function in Python that yields the powers of 2 up to a given exponent.

#Ans.6
def powers_of_2(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# Example usage:
max_exponent = 10
for power in powers_of_2(max_exponent):
    print(power)
```

```python
#Ques.9  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

#Ans.9
# Define a function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40, 50]

# Use map() to apply the conversion function to each element in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the list of temperatures in Fahrenheit
print("Temperatures in Celsius:", celsius_temperatures)
print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```

```python
#Ques.10  Create a Python program that uses `filter()` to remove all the vowels from a given string.

#Ans.10
# Define a function to check if a character is a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# The input string
input_string = "Hello, World!"

# Use filter() to remove vowels from the input string
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Print the resulting string
print("Original string:", input_string)
print("String without vowels:", filtered_string)
```
